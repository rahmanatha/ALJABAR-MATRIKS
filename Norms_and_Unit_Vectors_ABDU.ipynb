```python
import numpy as np
```

```python
# buat vector x dengan ukuran 3x1
x = np.array([[1],
              [2],
              [3]])
```

```python
# hitung nilai L2 norm dari vector x
norm_x = np.linalg.norm(x)
```

```python
# buat vector v dengan elemen (3,4)
v = np.array([3, 4])
```

```python
# hitung nilai L2 norm dari vector v
norm_v = np.linalg.norm(v)
```

```python
# tampilkan unit vector dari vector v
unit_v = v / norm_v
```

```python
# hitung panjang unit vector dari vector v
panjang_unit_v = np.linalg.norm(unit_v)
print("Vector x:\n", x)
print("L2 norm x:", norm_x)

print("\nVector v:", v)
print("L2 norm v:", norm_v)

print("\nUnit vector v:", unit_v)
print("Panjang unit vector v:", panjang_unit_v)
```
